```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('../../Data/Spiral/explode-rotated.npy')

random_mat = np.random.rand(10, 10)
sparsity_matrix = np.random.uniform(0, 1, (10, 10))

sparsity_th = 0
sparsity_matrix = sparsity_matrix > sparsity_th

random_mat = random_mat * sparsity_matrix
```

```python
data.shape
```

```python
random_mat
```

```python
for i in range(10):
    plt.plot(np.arange(1000), data[0, i, :], 'o')
```

```python
shape_0 = data.shape[0]
shape_1 = data.shape[1]
shape_2 = data.shape[2]

data_transform = np.zeros((shape_0, shape_1, shape_2))
data_transform.shape
```

```python
for i in range(data.shape[0]):
    data_transform[i, :, :] = random_mat @ data[i, :, :]

data_transform.shape
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
for i in range(6):
    ax[0].plot(np.arange(1000), data_transform[i, 0, :])
    
for i in range(6):
    ax[1].plot(np.arange(1000), data[i, 0, :])
```

```python
plt.plot(data[0, 0, :], data[0, 1, :])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
for i in range(10):
    ax[0].plot(np.arange(1000), data_transform[0, i, :])
    
for i in range(10):
    ax[1].plot(np.arange(1000), data[0, i, :])
```
