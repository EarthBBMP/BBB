```python
import numpy as np
import matplotlib.pyplot as plt
```

# Goal 
To generate data (6 distinct groups) with trajectory in time

data_structure = [[n_1 data], [n_2 data]]

```python
# c = number of class
# t = timepoints
# d = dimention of data

c = 6
t = 1000
d = 2
a = 1
```

```python
theta = np.linspace(0, 2*np.pi, c+1)
theta = theta[:-1]
theta
```

```python
def find_coords(theta, class_num):
    x = a * np.cos(theta[class_num])
    y = a * np.sin(theta[class_num])
    return [x, y]
```

```python
data = np.zeros((c, d, t))
```

```python
# Initial location
for i in range(c):
    data[i, :, 0] = find_coords(theta, i)
    print(find_coords(theta, i))    
```

```python
for i in range(t):
    if i == 0:
        continue
    data[:, :, i] = data[:, :, i-1] + np.random.normal(0, 0.01, (c, d))
```

```python
ax = plt.figure().add_subplot(projection='3d')

for i in range(c):
    ax.plot(data[i, 0, :], data[i, 1, :], range(t))

ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('Time')

```
